    Soru 1) Gradient Descent algortimasını kullanarak x^2+2*x+1 fonksiyonu için gloabil minumum noktasını bulunuz.
    Soru 2) Learning rate parametresini 0.001, 0.01,0.0001,0.1 olarak deneyerek sonuçlarını karşılaştırınız.

```python
def my_func(x):
  return (x**2)+(2*x)+1
```

```python
import numpy as np
my_linspace = np.linspace(-5,5,50)

import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
plt.plot(my_linspace,my_func(my_linspace))
plt.title("My Func")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

```python
def my_fl(x):
  return (2*x)+2
```

```python
def gradient_descent(x, my_fl, learn_rate, max_iter, tol=1e-3):
  next_point = tol+1
  logs = [x]
  while next_point > tol:
    for i in range(max_iter):
      next_point = learn_rate*my_fl(x) 
      x -= next_point
      logs.append(x)
  
  return logs
```

```python
learning_rates=[0.001, 0.01,0.0001,0.1]
```

```python
a = gradient_descent((my_linspace[-1]), my_fl, learning_rates[-1], 999)
```

```python
b=list()
for i in a:
  b.append(my_func(i))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,7))
font1 = {'family':'serif','color':'blue','size':20}
plt.plot(my_linspace,my_func(my_linspace), color="green", linewidth=5, label="My Function")
plt.plot(a, b, color="red",marker="o",ms = 10, label="GD")
plt.title(f"For Learning Rate: {learning_rates[-1]}",fontdict = font1)
plt.legend(fontsize=20)
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

```python
a = list()
for i in learning_rates:
  a.append(gradient_descent((my_linspace[-1]), my_fl, i, 999)[-1])
```

```python
import pandas as pd
a = pd.DataFrame(data=a,columns=["Global Minimum"])
b = pd.DataFrame(data=learning_rates,columns=["Learning Rates"])
a = pd.concat([a,b],axis=1)
a.set_index("Learning Rates")
```
