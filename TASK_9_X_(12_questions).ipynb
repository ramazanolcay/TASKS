    3) Feature map veya bir sonraki layerın boyutlarını belirleyen özellikler nelerdir. Açıklayınız.

Filter (Kernel): Bu katmanda girdi olarak gelen resim bir filtreden geçirilir. Örneğin, 5x5 input olan bir matrise 3x3 filtre uygularsak ve adım sayısı (stride) 1 olduğunu düşünürsek çıktısı  3x3 olur.

Pooling: Havuzlama olarak adlandırabileceğimiz bir işlemdir. Maksimum, minimum, ortalama gibi farklı versiyonları vardır ve örneğin, 5x5 bir matrise 2x2 pooling işlemi uyguladığımızda (stride =1) 3x3 şeklinde bir çıktı elde ederiz.

Padding: Filtreleme veya havuzlama işlemlerinde boyut kaybı olduğu için kullanılan bir işlemdir. 5x5 bir matrise zero padding uyguladığımızda 6x6 olarak elde ederiz ve 3x3 filtreleme (stride=1) sonrasında yine 5x5 bir matris elde etmiş oluruz.
