```python
def prime_checker(num):
  if num <3:
    if (num == 2):
      return True
    elif num == 1:
      return False
    else:
      print("There is something wrong")
  elif num>=3:
    if num%2 == 0:
      return False
    else:
      i=3
      while i < int(num/2 + 1):
        if num%i == 0:
          return False
        else:
          i += 2
      return True

```

```python
for i in range(1,100):
  print("Is ", i , " prime: ", prime_checker(i))
```
