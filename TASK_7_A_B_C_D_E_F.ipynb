TASK-7F: Then please develop a machine learning model with the 3 derived features that you obtain in 7E via the use of train-test split (be careful about stratify parameter in this function) and gridsearchCV function as well as a classification algorithm (such as Gaussian Naive Bayes) but please do not forget to optimize its hyperparameters. Finally, please report your findings via the use of classification report function (also a sklearn function). 


```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data.iloc[:,-3:], data["CDRGLOB"], test_size=0.20, random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
gnb = KNeighborsClassifier()
gnb.fit(X_train,y_train)
gnb_predict = gnb.predict(X_test)

from sklearn.metrics import accuracy_score
accuracy_score(y_test, gnb_predict)
```

```python
my_n_neighbors = my_n_estimators = np.linspace(1,15,num=5,dtype="int")
print(my_n_neighbors)

my_weights=['uniform','distance']

my_algorithm = ["auto", "ball_tree", "kd_tree", "brute"]

my_leaf_size = np.linspace(10,50,num=5,dtype="int")
print(my_leaf_size)

my_p = np.linspace(1,5,num=5,dtype="int")
print(my_p)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'n_neighbors': my_n_neighbors,
              'weights' : my_weights,
              'algorithm' : my_algorithm,
              'leaf_size' : my_leaf_size,
              'p' : my_p
              }

grid = GridSearchCV(estimator=gnb, param_grid=param_grid)
```

```python
grid.fit(X_train, y_train);
```

```python
grid.best_params_
#Accuracy changed inversely with var_smoothing value,
```

```python
gnb = grid.best_estimator_
my_predictions = gnb.fit(X_train,y_train).predict(X_test)
```

```python
from sklearn.metrics import classification_report
target_names = ["1","2","3","4"]
print(classification_report(y_test, my_predictions, target_names=target_names))
```
